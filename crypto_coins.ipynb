```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
sns.set_context('talk')
```

```python
def create_df(filename):
    return pd.read_csv(filename)
```

```python
# LOAD DATA
```

```python
btc = create_df('BTC-USD-2.csv')
doge = create_df('DOGE-USD-2.csv')
eth = create_df('ETH-USD-2.csv')
```

```python
# REMOVE NULL DATA
```

```python
btc.dropna(how='any', inplace=True)
doge.dropna(how='any', inplace=True)
eth.dropna(how='any', inplace=True)
```

```python
# SORT BY DATE (JUST TO MAKE SURE)
```

```python
btc = btc.sort_values(by='Date')
doge = doge.sort_values(by='Date')
eth = eth.sort_values(by='Date')
```

```python
xticks = []
# for yr in range(2020, 2022):
#     for month in range(1, 13):
#         xticks.append(str(yr) + "-" + str(month).rjust(2, '0') + "-01")
#         xticks.append(str(yr) + "-" + str(month).rjust(2, '0') + "-16")
for i in range(6, 13):
#     xticks.append(
#         "2020-" + str(i).rjust(2, '0') + '-01'
#     )
    xticks.append(
        "2020-" + str(i).rjust(2, '0') + '-16'
    )
    
for i in range(1, 7):
#     xticks.append(
#         "2021-" + str(i).rjust(2, '0') + '-01'
#     )
    xticks.append(
        "2021-" + str(i).rjust(2, '0') + '-16'
    )
```

```python
def plot_me(x, y, title, width=80, length=10):
    plt.figure(figsize=(width, length))
    plt.plot(btc[x], btc[y], 'r-', label='BTC')
    plt.plot(doge[x], doge[y], 'g-', label='Doge')
    plt.plot(eth[x], eth[y], 'b-', label='ETH')
    plt.legend()
    plt.title(title)
    plt.xticks(xticks, rotation=90)
    plt.savefig(y + '_plot.jpg')
    plt.show()
```

```python
for i in btc.columns:
    if i != 'Date':
        plot_me('Date', i, i + ' Plot')
```

```python
def plot_me2(df, title, filename):
    plt.figure(figsize=(80,10))
    plt.plot(df['Date'], df['Open'], 'r-', label='Open')
    plt.plot(df['Date'], df['Close'], 'g-', label='Close')
    plt.plot(df['Date'], df['Adj Close'], 'b-', label='Low')
    plt.plot(df['Date'], df['High'], 'c-', label='High')
    plt.plot(df['Date'], df['Low'], 'y-', label='Low')
    plt.legend()
    plt.title(title)
    plt.xticks(xticks, rotation=90)
    plt.savefig(filename + '.jpg')
    plt.show()
```

```python
plot_me2(btc, "BTC Details", 'btc')
plot_me2(doge, "Doge Details", 'doge')
plot_me2(doge, "ETH Details", 'eth')
```
